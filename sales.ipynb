# Description des jeux de données

### Advance Retail Sales: Clothing and Clothing Accessory Stores
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `RSCCAS`: Ventes au détail dans les magasins de vêtements et d'accessoires de vêtements.
- **Période**: De janvier 1992 à décembre 2024

### Advance Retail Sales: Gasoline Stations
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `RSGASS`: Ventes au détail dans les stations-service.
- **Période**: De janvier 1992 à décembre 2024

### Retail Sales: Electronics and Appliance Stores
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `MRTSSM443USS`: Ventes au détail dans les magasins d'électronique et d'appareils électroménagers.
- **Période**: De janvier 1992 à décembre 2024

### Retail Sales: Hobby, Toy, and Game Stores
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `MRTSSM45112USN`: Ventes au détail dans les magasins de loisirs, de jouets et de jeux.
- **Période**: De janvier 1992 à décembre 2024

### Retail Sales: Jewelry Stores
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `MRTSSM44831USS`: Ventes au détail dans les bijouteries.
- **Période**: De janvier 1992 à février 2024

### CPI Data
- **Fréquence**: Annuelle
- **Colonnes**:
  - `Year`: L'année de l'observation.
  - `Annual Average CPI(-U)`: Indice des prix à la consommation moyen annuel.
  - `Annual Percent Change (rate of inflation)`: Taux de variation annuel (taux d'inflation).
- **Période**: De 1914 à 2024

### Federal Funds Effective Rate
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `FEDFUNDS`: Taux effectif des fonds fédéraux.
- **Période**: De juillet 1954 à janvier 2025

### Fixed Rate Mortgage Average United States
- **Fréquence**: Hebdomadaire
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `MORTGAGE30US`: Taux hypothécaire fixe moyen sur 30 ans aux États-Unis.
- **Période**: De avril 1971 à juillet 2024

### Fred Unemployment Rate
- **Fréquence**: Mensuelle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `UNRATE`: Taux de chômage.
- **Période**: De janvier 1948 à janvier 2025

### GDP
- **Fréquence**: Trimestrielle
- **Colonnes**:
  - `observation_date`: La date de l'observation.
  - `GDP`: Produit intérieur brut (PIB).
- **Période**: De janvier 1947 à octobre 2024

```python
import pandas as pd
```

```python
clothing_sales = pd.read_csv('data/Advance Retail Sales: Clothing and Clothing Accessory Stores.csv')
gasoline_sales = pd.read_csv('data/Advance Retail Sales: Gasoline Stations.csv')
electronics_sales = pd.read_csv('data/Retail Sales: Electronics and Appliance Stores.csv')
games_sales = pd.read_csv('data/Retail Sales: Hobby, Toy, and Game Stores.csv')
jewelry_sales = pd.read_csv('data/Retail Sales: Jewelry Stores.csv')
restaurant_sales = pd.read_csv('data/Retail Sales: Restaurants and Other Eating Places.csv')
supermarket_sales = pd.read_csv('data/Retail Sales: Supermarkets and Other Grocery (Except Convenience) Stores.csv')
```

```python
# Renaming columns for better understanding
clothing_sales.rename(columns={'RSCCAS': 'Clothing_Sales'}, inplace=True)
gasoline_sales.rename(columns={'RSGASS': 'Gasoline_Sales'}, inplace=True)
electronics_sales.rename(columns={'MRTSSM443USS': 'Electronics_Sales'}, inplace=True)
games_sales.rename(columns={'MRTSSM45112USN': 'Games_Sales'}, inplace=True)
jewelry_sales.rename(columns={'MRTSSM44831USS': 'Jewelry_Sales'}, inplace=True)
restaurant_sales.rename(columns={'MRTSSM7225USN': 'Restaurant_Sales'}, inplace=True)
supermarket_sales.rename(columns={'MRTSSM44511USN': 'Supermarket_Sales'}, inplace=True)
```

```python
print("Clothing Sales:")
print(clothing_sales.head(), "\n")

print("Gasoline Sales:")
print(gasoline_sales.head(), "\n")

print("Electronics Sales:")
print(electronics_sales.head(), "\n")

print("Games Sales:") 
print(games_sales.head(), "\n")

print("Jewelry Sales:")
print(jewelry_sales.head(), "\n")

print("Restaurant Sales:")
print(restaurant_sales.head(), "\n")

print("Supermarket Sales:")
print(supermarket_sales.head())
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Configuration du style et de la taille de la figure
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(15, 10))

# Conversion des dates pour tous les dataframes
for df in [clothing_sales, gasoline_sales, electronics_sales, games_sales, 
           jewelry_sales, restaurant_sales, supermarket_sales]:
    df['observation_date'] = pd.to_datetime(df['observation_date'])

# Création du subplot
plt.subplot(111)

# Tracer toutes les séries
plt.plot(clothing_sales['observation_date'], clothing_sales['Clothing_Sales'], 
         label='Clothing', linewidth=2)
plt.plot(gasoline_sales['observation_date'], gasoline_sales['Gasoline_Sales'], 
         label='Gasoline', linewidth=2)
plt.plot(electronics_sales['observation_date'], electronics_sales['Electronics_Sales'], 
         label='Electronics', linewidth=2)
plt.plot(games_sales['observation_date'], games_sales['Games_Sales'], 
         label='Games', linewidth=2)
plt.plot(jewelry_sales['observation_date'], jewelry_sales['Jewelry_Sales'], 
         label='Jewelry', linewidth=2)
plt.plot(restaurant_sales['observation_date'], restaurant_sales['Restaurant_Sales'], 
         label='Restaurants', linewidth=2)
plt.plot(supermarket_sales['observation_date'], supermarket_sales['Supermarket_Sales'], 
         label='Supermarkets', linewidth=2)

# Personnalisation du graphique
plt.title('Évolution des ventes au détail par secteur', fontsize=14, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Ventes (en millions de dollars)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# Rotation des dates pour une meilleure lisibilité
plt.xticks(rotation=45)

# Ajustement automatique de la mise en page
plt.tight_layout()
```

```python
# Load Federal Funds data
fed_funds = pd.read_csv('data/Federal_Funds_Effective_Rate.csv')
fed_funds['observation_date'] = pd.to_datetime(fed_funds['observation_date'])
fed_funds.rename(columns={'FEDFUNDS': 'Federal_Funds_Rate'}, inplace=True)

# Create the plot
plt.figure(figsize=(15, 8))
plt.plot(fed_funds['observation_date'], fed_funds['Federal_Funds_Rate'], linewidth=2, color='blue')

plt.title('Federal Funds Effective Rate Over Time', fontsize=14, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)

plt.tight_layout()
```

```python
from plotly.subplots import make_subplots

import plotly.graph_objects as go

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Function to standardize data
def standardize(data):
    return (data - data.mean()) / data.std()

# Filter data from 1990
start_date = pd.to_datetime('1990-01-01')

# Plot retail data on primary y-axis
for df, label, color in [
    (clothing_sales, 'Clothing', '#1f77b4'),
    (gasoline_sales, 'Gasoline', '#ff7f0e'),
    (electronics_sales, 'Electronics', '#2ca02c'),
    (jewelry_sales, 'Jewelry', '#d62728'),
    (restaurant_sales, 'Restaurants', '#9467bd'),
    (supermarket_sales, 'Supermarkets', '#8c564b')
]:
    mask = df['observation_date'] >= start_date
    fig.add_trace(
        go.Scatter(
            x=df.loc[mask, 'observation_date'],
            y=standardize(df.loc[mask, df.columns[1]]),
            name=label,
            line=dict(width=2)
        ),
        secondary_y=False
    )

# Plot Fed Funds rate on secondary y-axis
mask_fed = fed_funds['observation_date'] >= start_date
fig.add_trace(
    go.Scatter(
        x=fed_funds.loc[mask_fed, 'observation_date'],
        y=standardize(fed_funds.loc[mask_fed, 'Federal_Funds_Rate']),
        name='Fed Funds Rate',
        line=dict(width=2, dash='dash', color='black')
    ),
    secondary_y=True
)

# Update layout
fig.update_layout(
    title='Evolution of Standardized Retail Sales and Fed Funds Rate (from 1990)',
    xaxis_title='Date',
    yaxis_title='Standardized Retail Sales',
    yaxis2_title='Standardized Fed Funds Rate',
    hovermode='x unified',
    showlegend=True,
    template='plotly_white',
    width=1200,
    height=700
)

fig.show()
```

```python
# Create a list of all dataframes and their column names
data_pairs = [
    (clothing_sales, 'Clothing_Sales'),
    (gasoline_sales, 'Gasoline_Sales'),
    (electronics_sales, 'Electronics_Sales'),
    (games_sales, 'Games_Sales'), 
    (jewelry_sales, 'Jewelry_Sales'),
    (restaurant_sales, 'Restaurant_Sales'),
    (supermarket_sales, 'Supermarket_Sales'),
    (fed_funds, 'Federal_Funds_Rate')
]

# Merge all dataframes on observation_date
merged_df = data_pairs[0][0][['observation_date', data_pairs[0][1]]].copy()
for df, col in data_pairs[1:]:
    merged_df = pd.merge(merged_df, df[['observation_date', col]], on='observation_date', how='outer')

# Calculate correlation matrix
correlation_matrix = merged_df.drop('observation_date', axis=1).corr()

# Create heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            center=0,
            fmt='.2f',
            square=True)

plt.title('Correlation Matrix of Retail Sales and Fed Funds Rate', pad=20)
plt.tight_layout()
```
